```python
import pandas as pd
```

```python
temposCorte = pd.DataFrame(page.get_all_records())
```

```python
temposCorte.head()
```

```python
### Ao rodar, irá retirar os simbolos de "(", ")" e "/", álem de substituir os " " por "_".
temposCorte.columns = temposCorte.columns.str.replace('(', '')
temposCorte.columns = temposCorte.columns.str.replace(')', '')
temposCorte.columns = temposCorte.columns.str.replace('/', '')
temposCorte.columns = temposCorte.columns.str.replace(' ', '_')
```
